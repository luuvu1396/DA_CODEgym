```python
## Gom nhóm đơn biến
# Với mỗi cửa hàng hãy tính trung bình cộng số người theo dõi
df.groupby('name')['follower_count'].mean()
# dataFrame.groupby(‘Thuộc tính cần phân nhóm’)[‘Thuộc tính cần tính giá trị’].phép toán()
```
