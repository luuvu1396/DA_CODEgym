```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/shopeep_koreantop_clothing_shop_data.csv')
df.head()
```

```python
# Để đơn giản ta thực hiện lọc tập dữ liệu ban đầu theo các thuộc tính sau
df = df[['date_collected','shop_location','response_time']]
df
```

```python
# 2. Tách cột dữ liệu có kiểu chuỗi
# Ta có thể sử dụng các phép toán tách chuỗi split
# Hàm str.split(‘Ký tự phân cách’) có tác dụng tách chuỗi ban đầu (Series.str.split(pat=None, *, n=- 1, expand=False, regex=None))
# str thành các chuỗi con được ngăn cách nhau bởi ký tự phân cách

# Tách cột shop_location thành 2 cột District và City


df['District']=df['shop_location'].str.split(',').str[0]
# df['District'] => tạo cột
# df['shop_location'] => lấy dữ liệu từ cột shop_location
# str.split(',') => tách dữ liệu ngăn cách bởi dấu phẩy
# str[0] => lấy giá trị của index đầu tiên
df['City']=df['shop_location'].str.split(',').str[1]
df
```

```python
# 3. Tách cột dữ liệu có kiểu ngày tháng
# Tách cột date_collected thành 3 cột Day, Month, Year
# pd.to_datetime(your_date_data, format="Your_datetime_format")
# Series.dt.day
df['Day']=pd.to_datetime(df['date_collected'],format='%Y-%m-%d').dt.day
# df['Day'] => tạo cột ngày
# pd.to_datetime(df['date_collected'],format='%Y-%m-%d') => định dạng datime
# .dt.day => lấy giá trị ngày
df['Month']=pd.to_datetime(df['date_collected'],format='%Y-%m-%d').dt.month
df['Year']=pd.to_datetime(df['date_collected'],format='%Y-%m-%d').dt.year
df
```

```python
# 4. Tách cột dữ liệu có kiểu thời gian
# Tách cột response_time thành 3 cột Hour, Minute, Second

df['Hour']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.hour
df['Minute']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.minute
df['Second']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.second
df
```
