```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
```

```python
df = pd.read_csv("/content/FoodPrice_in_Turkey (1).csv")
df
```

```python
# Vẽ các biểu đồ cột so sánh giá Milk (powder, infant formula) và Fuel (gas) tháng 12 cuối năm năm 2019
# của Ankara, Istanbul, Izmir và National Average.
```

```python
df["ProductName"].unique()
```

```python
df1 = df[(df["ProductName"] == "Milk (powder, infant formula) - Retail") & (df["Month"] == 12) & (df["Year"] == 2019)].reset_index()
df2  = df[(df["ProductName"] == "Fuel (gas) - Retail") & (df["Month"] == 12) & (df["Year"] == 2019)].reset_index()
df2
```

```python
# speed = [0.1, 17.5, 40, 48, 52, 69, 88]
# lifespan = [2, 8, 70, 1.5, 25, 12, 28]
# index = ['snail', 'pig', 'elephant',
#          'rabbit', 'giraffe', 'coyote', 'horse']
# df = pd.DataFrame({'speed': speed,
#                    'lifespan': lifespan}, index=index)
# df
# # ax = df.plot.bar(rot=0)
```

```python
df3 = pd.DataFrame({"Place1": df1["Place"], "Milk (powder, infant formula) - Retailspeed": df1["Price"],
                   "Fuel (gas) - Retail": df2["Price"]})
df3
```

```python
df3.set_index('Place1',inplace = True)
```

```python
ax = df3.plot.bar(rot=0)
plt.title('Milk and Gas Price in 12/2019', fontsize = 16, color = 'r')
plt.xlabel('Place', fontsize = 14)
plt.ylabel('Price', fontsize = 14)
plt.show()
```

```python
# "Vẽ các biểu đồ đường" phân tích xu hướng giá gạo (Rice-Retail), giá Fuel (gas)
# trung bình cả nước (National Average)
# trong năm 2016, 2018, 2019 tại Thổ Nhĩ Kì.
```

```python
df4 = df[(df['ProductName'] == 'Rice - Retail') & (df['Place'] == "National Average")
          & (df['Year'].isin([2016,2018,2019]))].reset_index()
df5 = df[(df['ProductName'] == 'Fuel (gas) - Retail') & (df['Place'] == "National Average")
          & (df['Year'].isin([2016,2018,2019]))].reset_index()
df5
```

```python
df6 = pd.DataFrame({"Year": df4["Year"],"Month": df4["Month"], "Rice": df4["Price"],
                   "Fuel (gas)": df5["Price"]})
df6.head()
```

```python
group = df6.groupby(["Year"])["Rice - Retail","Fuel (gas) - Retail"]
group
```

```python
lines = group.plot.line( marker = 'o', xlabel = 'Month', ylabel = 'Price')
plt.show()
```

```python
scatters = group.plot.scatter("Rice - Retail","Fuel (gas) - Retail",xlabel = 'Price of Rice - Retail',
                              ylabel = 'Price of Fuel (gas) - Retail')
plt.show()
```

```python
df11 = df6[df6["Year"] == 2016]
df12 = df6[df6["Year"] == 2018]
df13 = df6[df6["Year"] == 2019]
df11
```

```python
fig, ax = plt.subplots(3)
ax[0].plot(df11['Month'], df11['Rice'], marker = '*', label = 'Rice-2016')
ax[0].plot(df11['Month'], df11['Fuel (gas)'], marker = 's', label = 'Gas-2016')
ax[0].set_ylabel('Price')
ax[0].set_xticklabels([])
ax[0].set_title('2016')

ax[1].plot(df12['Month'], df12['Rice'], marker = '*', label = 'Rice-2018')
ax[1].plot(df12['Month'], df12['Fuel (gas)'], marker = 's', label = 'Gas-2018')
ax[1].set_ylabel('Price')
ax[1].set_xticklabels([])
ax[1].set_title('2018')

ax[2].plot(df13['Month'], df13['Rice'], marker = '*', label = 'Rice')
ax[2].plot(df13['Month'], df13['Fuel (gas)'], marker = 's', label = 'Gas')
ax[2].set_ylabel('Price')
ax[2].set_xlabel('Month')
ax[2].legend()
ax[2].set_title('2019')

plt.show()
```

```python
# Vẽ biểu đồ Scatter phân tích mối liên quan giữa giá gạo và giá gas
# trung bình quốc gia (National Average) tại Thổ Nhĩ Kì
# các năm 2016, 2018, 2019.
```

```python
fig, ax = plt.subplots(3)
ax[0].scatter(df11['Rice'], df11['Fuel (gas)'])
ax[0].set_title('2016')
ax[0].set_ylabel('Fuel (gas)')
ax[0].set_xticklabels([])

ax[1].scatter(df12['Rice'], df12['Fuel (gas)'])
ax[1].set_title('2018')
ax[1].set_ylabel('Fuel (gas)')
ax[1].set_xticklabels([])

ax[2].scatter(df13['Rice'], df13['Fuel (gas)'])
ax[2].set_title('2019')
ax[2].set_ylabel('Fuel (gas)')
ax[2].set_xlabel('Rice')
plt.show()
```
