```python
df3=df.filter(['join_month','join_day','join_year'])
df3
```

```python
df4 = df[['join_month','join_day','join_year']]
df4
```

```python
#Cách 1: dùng phương thức filter, cho phép lấy các bản ghi rời rạc, các chỉ 
#số được ngăn cách nhau bởi dấu phẩy.

df5 = df3.filter([1,2,3,4,5,6,7],axis=0)
df5
```

```python
#Cách 2: truyền chỉ số trực tiếp, cho phép lấy một hoặc nhiều bản ghi liên tiếp
df6 = df3[0:8]
df6
```

```python
#Lấy thông tin kết hợp theo hàng và cột
df7 = df.filter(['pk_shop','rating_bad','rating_good'])[:8]
df7
```

```python
df8=df[['pk_shop','rating_bad','rating_good']][:8]
df8
```

```python
#Lọc bản ghi theo điều kiện
#TH1: Lọc các bản ghi theo điều kiện đơn
# Ví dụ: Tìm các bản ghi có join_month là ‘April’

df[df['join_month']=='April']
```

```python
# TH2: Lọc các bản ghi theo điều kiện phức
# Ví dụ: Tìm các cửa hàng tham gia vào năm 2021 và có số lượng theo dõi >= 10000
# df[(đk1) & (đk2)]
df[(df['join_year'] == 2021) & (df['follower_count'] >= 10000)]
```

```python
# TH3: Lọc các cột và các bản ghi theo điều kiện
# Hiển thị các thông tin về response_time, response_rate của các bản ghi có join_month là ‘April’
# Lọc df tháng tham gia = April, lấy cột repone_tỉm và response_rate
df[df['join_month']=='April'].filter(['response_time', 'response_rate'])
```

```python
# Lọc bản ghi sử dụng phương thức query (Phương thức truy vấn query giúp câu lệnh chạy nhanh hơn,
# các điều kiện được nối với nhau bởi các phép toán logic như: and (và), or (hoặc), not (phủ định) 
# và được để trong một cặp dấu nháy kép.)
# Ví dụ: Lọc các cửa hàng tham gia vào năm 2021 và có số lượng theo dõi >= 10000

df.query('(join_year == 2021) and (follower_count >= 10000)')
```
