```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/FoodPrice_in_Turkey.csv")
df
```

```python
# Vẽ biểu đồ cột so sánh giá gạo (Rice-Retail) tháng 12 năm 2019
# của Ankara, Istanbul, Izmir và National Average.
target = df[(df["ProductName"] == "Rice - Retail") & (df["Month"] == 12) & (df["Year"] == 2019)]
target
```

```python
ax = target.plot.bar('Place','Price',width = 0.5)
plt.title('Rice Price in 12/2019', fontsize = 16, color = 'r')
plt.xlabel('Place', fontsize = 14)
plt.ylabel('Price', fontsize = 14)
plt.show()
```

```python
# Vẽ biểu đồ đường phân tích xu hướng giá gạo (Rice-Retail)
# trung bình cả nước (National Average) trong năm 2019 tại Thổ Nhĩ Kì.
target_2 = df[(df["ProductName"] == "Rice - Retail") & (df["Place"] == "National Average") & (df["Year"] == 2019)]
target_2
```

```python
lines = target_2.plot.line(x='Month', y='Price', linewidth = 2, marker = 'o',
                           markersize=10, markerfacecolor='red', markeredgecolor = 'blue',
                           markeredgewidth=2)
plt.title('Rice Price of National Average in 2019', fontsize = 16, color = 'r')
plt.xlabel('Month', fontsize = 14)
plt.ylabel('Price', fontsize = 14)
plt.show()
```

```python
# Vẽ biểu đồ Scatter phân tích mối liên quan giữa giá gạo và
# giá gas trung bình quốc gia (National Average) tại Thổ Nhĩ Kì.
target_3 = df[(df["ProductName"] == "Rice - Retail")
              & (df["Place"] == "National Average") & (df['Year'] == 2019)]
target_4 = df[(df["ProductName"] == "Fuel (gas) - Retail")
              & (df["Place"] == "National Average") & (df['Year'] == 2019)]
target_4
```

```python
plt.scatter(target_4['Price'],target_3['Price'], s = 50)
plt.title('Relationship between Rice Price and Gas Price', fontsize = 16, color = 'r')
plt.xlabel('Gas', fontsize = 14)
plt.ylabel('Rice', fontsize = 14)
plt.show()
```
